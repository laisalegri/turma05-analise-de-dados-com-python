## 2. Git e GitHub — Controle de Versão e Seu Repositório

> 📌 **Repositório da turma × seu repositório:**  
> `cfneves/turma05-analise-de-dados-com-python` é para você **consultar o material do curso**. O repositório que você cria abaixo é o **seu** — onde fica o código que você escreve.

### Os conceitos essenciais

| Conceito | O que é | Analogia |
|---|---|---|
| **Repositório** | Pasta do projeto versionada pelo Git | A caixa onde você guarda tudo |
| **Commit** | Um "ponto de salvamento" das mudanças, com mensagem | Fotografia + legenda do estado atual |
| **Branch** | Cópia paralela de trabalho sem afetar a versão principal | Rascunho antes de passar a limpo |
| **Push** | Enviar seus commits do computador para o GitHub | Salvar na nuvem |
| **Pull** | Baixar as atualizações do GitHub para o computador | Sincronizar o arquivo |
| **Clone** | Copiar um repositório do GitHub para o computador | Baixar o projeto completo |

### O fluxo do dia a dia — como enviar atualizações

Toda vez que você quiser enviar novas alterações para o GitHub — depois que o repositório já estiver clonado — **repita este fluxo** (não precisa clonar de novo):

```
1. Editar arquivo   →   2. git add   →   3. git commit   →   4. git push
```

```python
# Execute no TERMINAL do VS Code — um comando por vez
git add meu_script.py                         # "Git, presta atenção nesse arquivo"
git commit -m "Adiciona análise de vendas"    # "Git, salva isso no histórico"
git push origin main                          # "Git, manda para o GitHub"
```

> 💡 **Mensagem de commit: seja específico.** `"Update"` não diz nada. `"Corrige cálculo de média quando notas são zero"` conta a história — você mesmo vai agradecer em duas semanas.

### Opção 2 — Criar pelo terminal (para quem já tem uma pasta no computador)

Se você já tem arquivos no computador e quer começar a versionar:

1. Abra a pasta no VS Code e no terminal: `git init`
2. Primeiro commit: `git add .` → `git commit -m "Primeiro commit — início do projeto"`
3. Crie repositório vazio no GitHub (`github.com/new`) — **NÃO marque "Add a README file"**
4. Conecte e envie: `git remote add origin URL` → `git push -u origin main`

> ⚠️ O repositório no GitHub precisa estar **completamente vazio** para receber seu histórico local.

---

### Quando o Git pede ajuda: conflito de merge

Ocorre quando você e um colega editam o **mesmo arquivo, na mesma linha** e os dois fazem push. O Git marca o trecho assim:

```
<<<<<<< HEAD
total = preco * quantidade
=======
total = preco * quantidade * 1.1  # com imposto
>>>>>>> feature/calculo-imposto
```

Não é erro — é o Git pedindo sua decisão. No VS Code com GitLens instalado, aparecem botões **"Accept Current Change"** / **"Accept Incoming Change"** — clique no que quiser manter, salve e faça um novo commit.

---

### O que o Git não deve guardar: `.gitignore`

Crie um arquivo chamado `.gitignore` (com ponto no início, sem extensão) na raiz do projeto. No VS Code: `File → New File…` → salve como `.gitignore` → escolha **Plain Text**:

```
*.pyc
__pycache__/
.env
venv/
.DS_Store
```

> 💡 **Nunca coloque senhas ou tokens de API em arquivos versionados.** Mesmo apagando depois, o histórico do Git guarda tudo.

## 1. Listas

Uma lista guarda vários valores, em ordem, e pode ser alterada depois de criada (é **mutável**).

### 🔹 Exemplo 1 — Lista de compras

````python
compras = ["arroz", "feijão", "leite"]

print(compras[0])      # primeiro item
print(compras[-1])     # último item

compras.append("café")  # lista é mutável: posso adicionar
print(compras)
````

> ⚠️ **Veja como é um erro real do Python.** A lista `compras` tem 4 itens (índices 0 a 3, depois do `.append`). Se você pedir um índice que ela não tem, como `compras[10]`, o Python recusa e mostra `IndexError: list index out of range` — "índice fora do intervalo". Tente rodar `print(compras[10])` numa célula nova para ver o erro de verdade.

### 🔹 Exemplo 2 — Calculando a média de uma lista de notas

````python
notas = [7.5, 8.0, 6.5, 9.0, 5.5]

media = sum(notas) / len(notas)
print(f"Média da turma: {media}")
````

## 2. Tuplas

Uma tupla também guarda vários valores em ordem, mas **não pode ser alterada** depois de criada (é **imutável**) — útil para dados que nunca devem mudar.

### 🔹 Exemplo 1 — Coordenadas fixas

````python
coordenadas = (10, 20)

print(f"X = {coordenadas[0]}, Y = {coordenadas[1]}")

# coordenadas[0] = 99  # isso daria erro! tupla não pode ser alterada
````

> ⚠️ **Veja como é um erro real do Python.** Remova o `#` da última linha da célula acima e execute — você vai ver `TypeError: 'tuple' object does not support item assignment` ("o objeto tupla não permite atribuição de item"). É a imutabilidade da tupla sendo aplicada na prática, não só explicada em texto.

### 🔹 Exemplo 2 — Tupla para data de nascimento

````python
nascimento = (15, 3, 2005)   # dia, mês, ano

dia = nascimento[0]
mes = nascimento[1]
ano = nascimento[2]

print(f"Data de nascimento: {dia}/{mes}/{ano}")
````

## 3. Dicionários

Um dicionário guarda valores acessados por **chave** (um nome), não por posição.

### 🔹 Exemplo 1 — Cadastro de produto

````python
produto = {"nome": "Mouse", "preco": 45.0, "estoque": 12}

print(f"{produto['nome']} custa R$ {produto['preco']}")
print(f"Estoque disponível: {produto['estoque']} unidades")
````

### 🔹 Exemplo 2 — Atualizando um valor no dicionário

Ao contrário da tupla, o dicionário **permite alterar valores** depois de criado — só a chave não muda.

````python
aluno = {"nome": "Carlos", "nota": 7.5}

print(f"Nota atual: {aluno['nota']}")    # 7.5

aluno["nota"] = 9.0                      # dicionário é mutável — pode alterar
print(f"Nota corrigida: {aluno['nota']}")  # 9.0
````

## 4. Combinando tudo: o carrinho de compras

Um carrinho de compras é uma **lista de dicionários** — cada item tem produto, preço e quantidade.

````python
carrinho = [
    {"produto": "Caderno", "preco": 12.5, "quantidade": 2},
    {"produto": "Lápis", "preco": 2.0, "quantidade": 5},
    {"produto": "Borracha", "preco": 1.5, "quantidade": 3},
]

total = 0
for item in carrinho:
    subtotal = item["preco"] * item["quantidade"]
    total = total + subtotal
    print(f"{item['produto']}: {item['quantidade']}x R$ {item['preco']} = R$ {subtotal}")

print(f"\nTotal do carrinho: R$ {total}")
````
