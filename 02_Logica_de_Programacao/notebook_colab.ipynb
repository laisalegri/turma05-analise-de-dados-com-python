## 3. Operadores Relacionais e Lógicos

Os operadores da seção 2 produzem **números**. Os operadores desta seção são diferentes: produzem `True` ou `False` — e é esse `True`/`False` que vai permitir que seu programa **decida** algo, na seção 4.

Operadores **relacionais** comparam valores e sempre resultam em `True` ou `False`. Operadores **lógicos** combinam várias comparações.

```
==  !=        igual a / diferente de
>   <         maior que / menor que
>=  <=        maior ou igual / menor ou igual

and           verdadeiro só se AMBOS os lados forem verdadeiros
or            verdadeiro se PELO MENOS UM lado for verdadeiro
not           inverte o valor
```

## 4. Estruturas Condicionais (if / elif / else)

Lembra de comparar valores e obter `True`/`False` na seção 3? Agora você vai usar exatamente isso dentro de um `if`, para fazer seu programa escolher um caminho.

```python
if condicao_1:
    # executa se condicao_1 for True
elif condicao_2:
    # executa se condicao_1 for False e condicao_2 for True
else:
    # executa se nenhuma condição anterior for True
```

> 💡 O bloco de cada condição precisa estar **indentado** (4 espaços) — é a indentação que diz ao Python o que pertence a cada bloco.

### 🔹 Exemplo 1 (resolvido) — Classificando uma nota

> ⚠️ **Veja como é um erro real do Python.** Na célula abaixo, remova o `#` do início das duas linhas e execute — você vai ver um `SyntaxError: expected ':'`, porque o `if` ficou sem o `:` no final. Não é "quebrar o código": é o Python avisando exatamente onde está o problema. Leia sempre a **última linha** da mensagem de erro primeiro — ela quase sempre diz o que falta.

````python
# idade = 18
# if idade >= 18
#     print("Maior de idade")
````

````python
nota = 7.5

if nota >= 9:
    print("Excelente")
elif nota >= 7:
    print("Bom")
elif nota >= 5:
    print("Regular")
else:
    print("Insuficiente")
````

### 🔹 Exemplo 2 (resolvido) — Decidindo o acesso a um site

````python
idade = 15

if idade >= 18:
    print("Acesso liberado")
else:
    print("Acesso bloqueado: conteúdo para maiores de 18 anos")
````

### 🔹 Exemplo 3 (resolvido) — Combinando "and" dentro do if

Os Exemplos 1 e 2 testavam **uma** condição por vez. Você também pode usar os operadores lógicos da seção 3 (`and`, `or`) dentro da condição do `if`, para testar várias coisas ao mesmo tempo:

````python
numero = 15

if numero % 3 == 0 and numero % 5 == 0:
    print(f"{numero} é múltiplo de 3 e de 5 ao mesmo tempo")
elif numero % 3 == 0:
    print(f"{numero} é múltiplo de 3")
elif numero % 5 == 0:
    print(f"{numero} é múltiplo de 5")
else:
    print(f"{numero} não é múltiplo de 3 nem de 5")
````

> 💡 **Por que o `and` vem primeiro?** Se você checasse "múltiplo de 3" antes do "múltiplo de 3 e de 5", o programa pararia no primeiro `elif` verdadeiro e nunca chegaria a avisar que o número é múltiplo dos dois. **Ordem importa**: sempre comece pela condição mais específica.

### 🔹 Exemplo 4 (resolvido) — Calculando e depois classificando

Até aqui, toda condição comparava uma variável já pronta (`idade`, `nota`). Mas você também pode **calcular um valor novo** (com os operadores da seção 2) e só depois classificá-lo com `if`/`elif`/`else`:

````python
renda_mensal = 3000
gasto_planejado = 900

razao = gasto_planejado / renda_mensal   # calcula primeiro...

if razao <= 0.30:                        # ...e só depois classifica
    print("Gasto dentro do limite recomendado (até 30% da renda)")
elif razao <= 0.50:
    print("Gasto no limite de atenção (30% a 50% da renda)")
else:
    print("Gasto acima do recomendado (mais de 50% da renda)")
````

## 5. Estruturas de Repetição (for / while)

Com `if`/`elif`/`else` (seção 4) seu programa já decide um caminho — mas só uma vez. Agora você vai repetir uma ação várias vezes, sem copiar e colar o mesmo código.

Use `for` quando você já sabe quantas vezes quer repetir. Use `while` quando a repetição depende de uma condição que pode mudar.

### 🔹 Exemplo 1 (resolvido) — Contando com for

````python
for numero in range(1, 6):
    print(f"Contando: {numero}")
````

### 🔹 Exemplo 2 (resolvido) — Contagem regressiva com while

````python
contador = 5

while contador > 0:
    print(contador)
    contador = contador - 1

print("Lançamento!")
````
