## 1. Pandas: Series e DataFrame

### 🔹 Exemplo 1 (resolvido) — Criando um DataFrame

## 2. Explorando um DataFrame: describe, filtros e groupby

### 🔹 Exemplo 1 (resolvido) — describe()

```python
print(vendas.describe())
```

> ⚠️ **Veja como é um erro real do Python.** Tente acessar uma coluna com nome errado, como `vendas["preco_errado"]` — você vai ver `KeyError: 'preco_errado'`. Antes de selecionar uma coluna, confira os nomes exatos com `vendas.columns`.

```python
try:
    print(vendas["preco_errado"])
except KeyError as e:
    print(f"Erro: coluna não encontrada -> {e}")
```

### Selecionando por posição ou por rótulo: `.loc` e `.iloc`

`.loc` usa rótulos (nomes), `.iloc` usa posições (números). Os dois fazem a mesma coisa de jeitos diferentes — use `.loc` quando você sabe o nome da coluna, e `.iloc` quando só sabe a posição.

```python
print(vendas.loc[0, "produto"])      # .loc: pelo nome da coluna + posição da linha
print(vendas.iloc[0, 0])             # .iloc: só por posição numérica (linha 0, coluna 0)
```

### 🔹 Exemplo 2 (resolvido) — Filtro e groupby

```python
# Filtro: só produtos vendidos mais de 20 vezes
mais_vendidos = vendas[vendas["quantidade_vendida"] > 20]
print(mais_vendidos)

# groupby: preço médio por categoria
print(vendas.groupby("categoria")["preco"].mean())
```

### Como o groupby pensa por dentro: Dividir → Aplicar → Combinar

1. **Dividir (split):** o Pandas separa as linhas em grupos, um para cada valor de `categoria`.
2. **Aplicar (apply):** dentro de cada grupo separado, calcula a função pedida (ex.: `mean()` dos preços apenas daquele grupo).
3. **Combinar (combine):** junta o resultado de cada grupo numa única Series/tabela de resposta.

É exatamente isso que a Tabela Dinâmica do Excel faz por trás dos panos — o `groupby` só está explicitando os 3 passos acima.

### Ordenando o resultado: `sort_values()`

`sort_values()` ordena os valores do maior para o menor (ou o contrário) — útil para responder "qual categoria tem o maior preço médio?" sem precisar procurar manualmente o maior número na tabela.

```python
preco_medio_por_categoria = vendas.groupby("categoria")["preco"].mean()
print(preco_medio_por_categoria.sort_values(ascending=False))
```

## 3. NumPy: arrays e operações vetorizadas

### 🔹 Exemplo 1 (resolvido) — Array x lista

```python
import numpy as np

precos = np.array([12.5, 2.0, 150.0, 1.5, 3.5])
quantidades = np.array([30, 80, 5, 60, 45])

receita = precos * quantidades   # multiplicação elemento a elemento, sem for!
print(receita)
```

> ⚠️ **Veja como é um erro real do Python.** Multiplicar duas listas comuns (não arrays) não funciona como multiplicação elemento a elemento — veja na célula abaixo.

```python
try:
    lista1 = [12.5, 2.0, 150.0]
    lista2 = [30, 80, 5]
    print(lista1 * lista2)
except TypeError as e:
    print(f"Erro: {e}")
```

### 🔹 Exemplo 2 (resolvido) — Broadcasting e funções estatísticas

```python
precos_com_aumento = precos * 1.10   # aumenta 10% em todos de uma vez
print(precos_com_aumento)

print(f"Receita total: {receita.sum()}")
print(f"Preço médio: {precos.mean()}")
print(f"Maior preço: {precos.max()}")
```

> ⚠️ **Quando o broadcasting NÃO funciona:** arrays de tamanhos incompatíveis geram um erro real, não um resultado "estranho" — veja na célula abaixo.

```python
try:
    a = np.array([1, 2, 3])
    b = np.array([1, 2])
    print(a * b)
except ValueError as e:
    print(f"Erro: {e}")
```

### ✏️ Atividade Prática 3 — Sua vez de programar

```python
meus_precos = np.array([...])
minhas_quantidades = np.array([...])

minha_receita = meus_precos * minhas_quantidades
print(minha_receita)
print(f"Receita total: {minha_receita.sum()}")
```

## 4. Combinando tudo: receita por categoria

Adicionamos a coluna `receita` ao DataFrame usando uma operação vetorizada do próprio Pandas (que por trás usa NumPy).

```python
vendas["receita"] = vendas["preco"] * vendas["quantidade_vendida"]
print(vendas)

receita_por_categoria = vendas.groupby("categoria")["receita"].sum()
print(receita_por_categoria.sort_values(ascending=False))
```
