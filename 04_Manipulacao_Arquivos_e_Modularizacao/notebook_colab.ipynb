## 1. Arquivos: CSV e JSON

### 🔹 Exemplo 1 — Escrevendo e lendo um arquivo JSON

````python
import json

produto = {"nome": "Mouse", "preco": 45.0, "estoque": 12}

with open("produto.json", "w", encoding="utf-8") as arquivo:
    json.dump(produto, arquivo, ensure_ascii=False)

with open("produto.json", "r", encoding="utf-8") as arquivo:
    dados = json.load(arquivo)

print(f"{dados['nome']} custa R$ {dados['preco']}")
````

### 🔹 Exemplo 2 — Escrevendo uma lista de dicionários em CSV

````python
import csv

alunos = [
    {"nome": "Ana", "nota": 8.5},
    {"nome": "Bruno", "nota": 7.0},
    {"nome": "Carla", "nota": 9.0},
]

with open("alunos.csv", "w", newline="", encoding="utf-8") as arquivo:
    escritor = csv.DictWriter(arquivo, fieldnames=["nome", "nota"])
    escritor.writeheader()
    escritor.writerows(alunos)

with open("alunos.csv", "r", encoding="utf-8") as arquivo:
    print(arquivo.read())
````

É o erro mais comum de quem trabalha com arquivos (nome digitado errado, ou arquivo que ainda não existe). Você pode se proteger com `try`/`except` — veja a célula abaixo.

## 2. Datas e Expressões Regulares

### 🔹 Exemplo 1 — Calculando dias restantes com `datetime`

````python
from datetime import date

hoje = date.today()
fim_do_curso = date(2026, 8, 14)

dias_restantes = (fim_do_curso - hoje).days
print(f"Faltam {dias_restantes} dias para o fim do curso.")
````

### 🔹 Exemplo 2 — Validando um e-mail com `re`

````python
import re

email = "aluno@exemplo.com"
padrao = r"^[^\s@]+@[^\s@]+\.[^\s@]+$"

if re.match(padrao, email):
    print("E-mail parece válido!")
else:
    print("E-mail inválido.")
````

## 3. Funções: parâmetros, valores padrão e `return`

Olhe este código que calcula o total de 3 clientes diferentes:
```python
total_ana = 10 * 2 - 0
total_bruno = 25 * 1 - 5
total_carla = 8 * 4 - 2
```
A mesma conta (`preço × quantidade − desconto`) está escrita 3 vezes — se a regra mudar um dia, é fácil esquecer de corrigir uma das linhas. Uma função resolve isso: você escreve a regra **uma única vez**.

### 🔹 Exemplo 1 — Função com retorno

````python
def media(lista):
    return sum(lista) / len(lista)

notas = [7.5, 8.0, 6.5]
print(media(notas))
````

### 🔹 Exemplo 2 — Parâmetro opcional (valor padrão)

````python
def calcular_total(preco, quantidade, desconto=0):
    return preco * quantidade - desconto

print(calcular_total(10, 2))        # sem desconto: 20
print(calcular_total(10, 2, 5))     # com desconto: 15
````

## 4. Lambda e Módulos

### 🔹 Exemplo 1 — Função lambda

````python
media_lambda = lambda lista: sum(lista) / len(lista)

print(media_lambda([7.5, 8.0, 6.5]))
````

### 🔹 Exemplo 2 — Onde o lambda realmente brilha: dentro de `sorted()`/`min()`/`max()`

Guardar uma lambda numa variável (Exemplo 1) é raro na prática. O uso real é passar a lambda **direto** como argumento de outra função, para dizer "compare por *este* critério":

````python
produtos = [
    {"nome": "Mouse", "preco": 45.0},
    {"nome": "Teclado", "preco": 89.0},
    {"nome": "Borracha", "preco": 2.0},
]

mais_barato = min(produtos, key=lambda p: p["preco"])
print(f"O mais barato é {mais_barato['nome']}, por R$ {mais_barato['preco']}")
````

### 🔹 Exemplo 3 — Criando e usando um módulo próprio

No Colab, criamos o "módulo" como um arquivo de texto e importamos com `%%writefile` (truque específico do Colab/Jupyter).
