## 1. Matplotlib: Gráficos Básicos

Todo gráfico em Matplotlib começa com a mesma receita: `fig, ax = plt.subplots()` (a "folha" e os "eixos"), depois `ax.plot(...)`/`ax.bar(...)`/`ax.scatter(...)`, e por fim `plt.show()`.

> 💡 Por que `fig, ax` desde o início? Existe um jeito mais curto (`plt.plot(...)` direto), mas só funciona bem para 1 gráfico isolado. O padrão `fig, ax` é o que você vai usar sempre — inclusive com vários gráficos juntos (Seção 2).

### 🔹 Exemplo 1 (resolvido) — Gráfico de linha

```python
import matplotlib.pyplot as plt
import pandas as pd

vendas = pd.DataFrame({
    "mes": ["Jan", "Fev", "Mar", "Abr", "Mai"],
    "receita": [1200, 1500, 900, 1800, 2100],
})

fig, ax = plt.subplots()
ax.plot(vendas["mes"], vendas["receita"])
plt.show()
```

### 🔹 Exemplo 2 (resolvido) — Gráfico de barras

```python
vendas_categoria = pd.DataFrame({
    "categoria": ["Papelaria", "Acessórios", "Eletrônicos"],
    "receita_total": [3200, 1800, 4500],
})

fig, ax = plt.subplots()
ax.bar(vendas_categoria["categoria"], vendas_categoria["receita_total"])
plt.show()
```

### 🔹 Exemplo 3 (resolvido) — Gráfico de dispersão

```python
fig, ax = plt.subplots()
ax.scatter(vendas["receita"], [10, 14, 8, 18, 21])  # ex.: receita x nº de pedidos
plt.show()
```

💡 **Esqueceu de exibir o gráfico no Colab?** Se nada aparecer, confira se a célula termina com `plt.show()`.

| Tipo de gráfico | Quando usar |
|---|---|
| Linha (`ax.plot`) | Evolução ao longo do tempo |
| Barra (`ax.bar`) | Comparar categorias |
| Dispersão (`ax.scatter`) | Relação entre 2 variáveis numéricas |

## 2. Customização, Subplots e Exportação

### 🔹 Exemplo 1 (resolvido) — Customizando um gráfico

```python
fig, ax = plt.subplots()
ax.plot(vendas["mes"], vendas["receita"], color="#CC0000", marker="o", label="Receita")
ax.set_title("Receita mensal")
ax.set_xlabel("Mês")
ax.set_ylabel("Receita (R$)")
ax.legend()
plt.show()
```

### 🔹 Exemplo 2 (resolvido) — Subplots: vários gráficos lado a lado

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 linha, 2 colunas de gráficos

axs[0].plot(vendas["mes"], vendas["receita"])
axs[0].set_title("Evolução mensal")

axs[1].bar(vendas_categoria["categoria"], vendas_categoria["receita_total"])
axs[1].set_title("Por categoria")

plt.show()
```

### Exportação de gráficos

```python
fig.savefig("receita_mensal.png", dpi=150)
print("Gráfico exportado.")
```

## 3. Web Scraping

**Web scraping** é extrair dados diretamente de páginas web — útil quando o site não oferece uma API para isso.

> ⚠️ **Antes de fazer scraping de qualquer site real:** verifique o `robots.txt` e os termos de uso. Fazer muitas requisições rápidas pode sobrecarregar o servidor. Por isso, os exemplos abaixo usam o `quotes.toscrape.com` — um site público criado especificamente para prática de scraping.

### 🔹 Exemplo 1 (resolvido) — Buscando o HTML de uma página

```python
import requests

resposta = requests.get("http://quotes.toscrape.com/")
print(resposta.status_code)   # 200 = sucesso
print(resposta.text[:300])    # os primeiros 300 caracteres do HTML
```

> ⚠️ **Veja como é um erro real do Python.** Se você esquecer o `http://`/`https://` na URL: `requests.get("quotes.toscrape.com")` — você vai ver `MissingSchema: Invalid URL 'quotes.toscrape.com': No scheme supplied`.

```python
try:
    requests.get("quotes.toscrape.com")
except requests.exceptions.MissingSchema as e:
    print(f"Erro: {e}")
```

### 🔹 Exemplo 2 (resolvido) — Extraindo dados com BeautifulSoup

```python
from bs4 import BeautifulSoup

sopa = BeautifulSoup(resposta.text, "html.parser")
citacoes = sopa.find_all("span", class_="text")
autores = sopa.find_all("small", class_="author")

for citacao, autor in zip(citacoes[:3], autores[:3]):
    print(autor.text, "-", citacao.text)
```

### ✏️ Atividade Prática 3 — Sua vez de programar

Extraia as 5 primeiras tags `<span class="text">` de `quotes.toscrape.com` e guarde em uma lista.

```python
minhas_citacoes = [tag.text for tag in sopa.find_all("span", class_="text")[:5]]
print(minhas_citacoes)
```

## 4. Funções Reutilizáveis e Introdução a Classes

Você já cria funções desde a Semana 04. Mas há situações em que uma função não basta: quando você precisa agrupar **dados** e **comportamento** juntos.

### 🔹 Exemplo 1 (resolvido) — Função de transformação reutilizável

```python
def normalizar(coluna):
    return (coluna - coluna.min()) / (coluna.max() - coluna.min())

vendas["receita_normalizada"] = normalizar(vendas["receita"])
print(vendas)
```

### 🔹 Exemplo 2 (resolvido) — Sua primeira classe

```python
class Produto:
    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco

    def aplicar_desconto(self, percentual):
        self.preco = self.preco * (1 - percentual / 100)

mochila = Produto("Mochila", 150.0)
mochila.aplicar_desconto(10)
print(mochila.nome, mochila.preco)
```

| Termo | O que é |
|---|---|
| `class Produto:` | Define o "molde" — o que todo Produto vai ter/fazer |
| `__init__` | Roda automaticamente ao criar um Produto novo |
| `self` | Representa "este Produto específico" |
| `mochila = Produto(...)` | Cria um **objeto** — uma "cópia preenchida" do molde |

💡 **Função x Classe:** uma função recebe dados, processa, devolve um resultado, e esquece tudo depois. Uma classe **guarda** dados (no `self`) e empacota junto as ações que fazem sentido para aqueles dados.

> ⚠️ **Veja como é um erro real do Python.** Se você esquecer o `self` ao definir um método, ao chamar `Produto("Caderno", 12.5).mostrar()` você vai ver `TypeError: Produto.mostrar() takes 0 positional arguments but 1 was given`.

```python
class ProdutoComBug:
    def mostrar():  # faltou o self
        print("produto")

try:
    ProdutoComBug().mostrar()
except TypeError as e:
    print(f"Erro: {e}")
```

## 5. ETL e Integração Python-SQL

**ETL** significa **E**xtract (extrair) → **T**ransform (transformar) → **L**oad (carregar). Você já faz isso desde a Semana 04 — só não tinha esse nome:

| Etapa | O que você já aprendeu que se encaixa aqui |
|---|---|
| **E**xtract | Ler um arquivo (`pd.read_csv`) ou fazer scraping (Seção 3) |
| **T**ransform | Limpar e transformar com Pandas (Semana 06) |
| **L**oad | Salvar o resultado — em arquivo ou em banco de dados |

### 🔹 Exemplo 1 (resolvido) — Um pipeline ETL completo

```python
vendas.to_csv("vendas_bruto.csv", index=False)               # (preparando o exemplo)

dados = pd.read_csv("vendas_bruto.csv")                       # E - Extract
dados["receita_normalizada"] = normalizar(dados["receita"])  # T - Transform
dados.to_csv("vendas_tratado.csv", index=False)               # L - Load (em arquivo)
print("Pipeline concluído.")
```

### 🔹 Exemplo 2 (resolvido) — Load num banco de dados com sqlite3

```python
import sqlite3

conexao = sqlite3.connect("dados.db")
dados.to_sql("vendas", conexao, index=False, if_exists="replace")  # L - Load (em banco)

resultado = pd.read_sql("SELECT * FROM vendas WHERE receita > 1000", conexao)
print(resultado)

conexao.close()
```

💡 **Por que `sqlite3` e não PostgreSQL aqui?** `sqlite3` não precisa de servidor nem configuração. No mercado, o mesmo `pandas.to_sql()`/`pd.read_sql()` funciona com PostgreSQL (via `SQLAlchemy`), só muda a conexão.

### ✏️ Atividade Prática 5 — Sua vez de programar

Monte um pipeline ETL com seus próprios dados: leia um CSV, transforme com uma função sua, e carregue numa tabela SQLite.

```python
meu_df.to_csv("meus_dados.csv", index=False)

meus_dados = pd.read_csv("meus_dados.csv")                       # E
meus_dados["valor_normalizado"] = normalizar(meus_dados["valor"]) # T

minha_conexao = sqlite3.connect("meu_banco.db")
meus_dados.to_sql("meus_dados", minha_conexao, index=False, if_exists="replace")  # L
print(pd.read_sql("SELECT * FROM meus_dados", minha_conexao))
minha_conexao.close()
```
