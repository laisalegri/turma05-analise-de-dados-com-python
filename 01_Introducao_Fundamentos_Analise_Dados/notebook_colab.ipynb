## 2. Dado, Informação, Conhecimento e Tipo

| Conceito | Definição |
|---|---|
| Dado | Registro bruto, sem contexto |
| Tipo | A natureza desse dado (número, texto, data...) |
| Informação | Dado organizado, com contexto |
| Conhecimento | Informação aplicada a uma decisão |

### 🔹 Exemplo 1 — Temperatura corporal

Veja como isso aparece em código: o **dado** vira uma **variável**, e o `print()` mostra a **informação** formatada.

> 🔍 **Onde fica a fronteira entre Dado e Informação?** Se eu só dissesse "`152`", sem dizer mais nada, isso ainda seria só **dado** — um número solto, sem contexto. Só vira **informação** quando dizemos o que ele representa (veja o Exemplo 2 abaixo: "vendemos 152 unidades hoje, 20% acima da média"). O número sozinho não muda — o que muda é o contexto que colocamos em volta dele.
>
> 💡 **O padrão por trás do "Conhecimento":** quase toda decisão segue a mesma receita: *informação + um critério de comparação (faixa, meta, limite) = conhecimento*. Guarde esse padrão — ele vai se repetir o curso inteiro.

````python
# DADO: um número guardado em uma variável
temperatura = 37

# INFORMAÇÃO: o dado organizado, com contexto, mostrado na tela
print(f"A temperatura do paciente é {temperatura}°C")

# CONHECIMENTO: a decisão, comparando a informação com um critério/faixa conhecida
# a faixa de temperatura corporal considerada normal é 36°C a 37,5°C
print("37°C está dentro da faixa normal (36°C–37,5°C) — não preciso agir.")
````

### 🔹 Exemplo 2 — Vendas do dia em uma loja

````python
# DADO: um número guardado em uma variável
vendas_hoje = 152

# INFORMAÇÃO: o dado organizado, com contexto
print(f"Vendemos {vendas_hoje} unidades hoje, 20% acima da média das últimas 4 semanas")

# CONHECIMENTO: a decisão, comparando a informação com a meta/expectativa da loja
print("Como estamos 20% acima da média, vamos repor o estoque antes do fim de semana.")
````

## 3. As Etapas da Análise de Dados

```
1. COLETA        → reunir os dados
2. LIMPEZA       → corrigir erros, remover duplicatas e nulos
3. ANÁLISE       → buscar padrões e causas
4. VISUALIZAÇÃO  → transformar números em gráficos
5. COMUNICAÇÃO   → apresentar o resultado e a recomendação
```

Essas etapas vão se repetir em **todas as semanas seguintes** do curso.

### 🔹 Exemplo 1 — Loja de roupas com queda de vendas

**Cenário:** Uma loja de roupas percebeu que as vendas caíram nos últimos 3 meses e quer entender o motivo.

````python
print("1. Coleta:", "exportar os dados de vendas dos últimos 6 meses do sistema da loja")
print("2. Limpeza:", "remover vendas duplicadas e corrigir preços negativos ou nulos")
print("3. Análise:", "comparar vendas por região, por vendedor e por mês")
print("4. Visualização:", "gráfico de linha mostrando a queda mês a mês por região")
print("5. Comunicação:", "as vendas caíram 18% na região Sul após a saída de um vendedor-chave")
````

### 🔹 Exemplo 2 — Streaming de vídeo com aumento de cancelamentos

````python
print("1. Coleta:", "reunir dados de assinaturas, cancelamentos e horas de uso da plataforma")
print("2. Limpeza:", "remover contas de teste e datas de cancelamento inconsistentes")
print("3. Análise:", "comparar tempo de uso de quem cancelou com o de quem permaneceu")
print("4. Visualização:", "gráfico de barras comparando horas assistidas por mês entre os dois grupos")
print("5. Comunicação:", "quem cancela assiste 70% menos conteúdo no 1º mês — melhorar recomendações")
````
